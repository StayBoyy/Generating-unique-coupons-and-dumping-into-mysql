#                               Generating different types of Data Visualization graphs
____

# 1 : Bar Graph 


```python

plt.figure(figsize=(17, 10))

# Ploted for 2015
plt.subplot(1, 2, 1)
df.plot(kind='bar', x='States/UTs', y='Rape - 2015', legend=False, ax=plt.gca())
plt.title('Rape Cases-2015')
plt.xlabel('States/UTs')
plt.ylabel('Number of Cases')
plt.xticks(rotation=90)

# Plot for 2016
plt.subplot(1, 2, 2)
df.plot(kind='bar', x='States/UTs', y='Rape - 2016', legend=False, ax=plt.gca())
plt.title('Rape Cases-2016')
plt.xlabel('States/UTs')
plt.ylabel('Number of Cases')
plt.xticks(rotation=90)

plt.tight_layout() # Prevents from overlapping the two plots
plt.show()
```

# 2 : Pie Chart 

#### Zoom for clear view

```python

# Pie chart for 2015 data
plt.figure(figsize=(50, 40))

plt.subplot(3, 1, 3)
plt.pie(df['Rape - 2015'], labels=df['States/UTs'], autopct='%1.1f%%', startangle=140) # Here autopct='%1.1f%% converts into % and then float value
plt.title('Rape Cases Distribution - 2015')

# Pie chart for 2016 data
plt.subplot(3, 2, 3)
plt.pie(df['Rape - 2016'], labels=df['States/UTs'], autopct='%1.1f%%', startangle=140)
plt.title('Rape Cases Distribution - 2016')

plt.tight_layout()
plt.show()
## ZOOM for clear view as it get merge for small values 
```

# 3 : Histogram 

```python
# Ploting a simple histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Rape - 2015'], bins=10, edgecolor='black', color='skyblue')

# Adding labels and title
plt.title('Distribution of Rape Cases Across States/UTs - 2015')
plt.xlabel('Number of Rape Cases')
plt.ylabel('Number of States/UTs')

# Displaying the plot 
plt.show()
# Here We are ploting it for the year -->> 2015 
```

# 4 : Line Plot 


```python

x = df["States/UTs"]
y = df["Rape - 2015"]

# Ploting with dark background style
plt.style.use("dark_background")
plt.figure(figsize=(16, 10))

# Simple line plot with markers
plt.plot(x, y, marker='o', color='cyan')

# Adding title and labels
plt.title('Rape Cases in 2015 by States/UTs')
plt.xlabel('States/UTs')
plt.ylabel('Number of Cases')

# Rotated x-axis labels for readability
plt.xticks(rotation=90)

# Adding value annotations directly on the line
for i in range(len(x)):
    plt.text(i, y[i], str(y[i]), ha='center', va='bottom', fontsize=10, color='white') # For clear understanding and matching with our DATASET

plt.show()
# Here We are ploting it for the year -->> 2015  
```

# 5 :  Scatter Plot 

```python

x = df["States/UTs"]
y = df["Rape - 2015"]

# Ploting with dark background style
plt.style.use("dark_background")
plt.figure(figsize=(16, 10))

# Scatter plot
plt.scatter(x, y, color='cyan', s=100)  # s controls the size of the points

# Adding title and labels
plt.title('Rape Cases in 2015 by States/UTs')
plt.xlabel('States/UTs')
plt.ylabel('Number of Cases')

# Rotated x-axis labels for readability
plt.xticks(rotation=90)

# Adding value annotations directly on the scatter points
for i in range(len(x)):
    plt.text(i, y[i], str(y[i]), ha='center', va='bottom', fontsize=10, color='white')

plt.show()
# Here We are ploting it for the year -->> 2015 
```

# 6 : Area Plot 

```python

x = df["States/UTs"]
y = df["Rape - 2015"]

# Ploting with dark background style
plt.style.use("dark_background")
plt.figure(figsize=(16, 10))

# Area chart
plt.fill_between(x, y, color='cyan', alpha=0.5)

# Adding title and labels
plt.title('Rape Cases in 2015 by States/UTs')
plt.xlabel('States/UTs')
plt.ylabel('Number of Cases')

# Rotated x-axis labels for readability
plt.xticks(rotation=90)

# Adding value annotations directly on the area chart
for i in range(len(x)):
    plt.text(i, y[i], str(y[i]), ha='center', va='bottom', fontsize=10, color='white')

plt.show()
# Here also We are ploting it for the year -->> 2015 
```
